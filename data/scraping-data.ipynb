```python
import subprocess
import pandas as pd
from datetime import datetime, timedelta
import os

# Tentukan tanggal awal dan tanggal akhir scraping
start_date = "2025-04-01"
end_date = "2025-05-20"
limit = 1000

# Konversi start_date dan end_date ke format datetime untuk manipulasi tanggal
start_dt = datetime.strptime(start_date, "%Y-%m-%d")
end_dt = datetime.strptime(end_date, "%Y-%m-%d")

# Buat list untuk menyimpan nama file hasil scraping
file_list = []

# Loop untuk setiap tanggal dalam rentang waktu yang ditentukan
current_dt = start_dt
while current_dt <= end_dt:
    # Format tanggal scraping saat ini
    since_date = current_dt.strftime("%Y-%m-%d")
    until_date = (current_dt + timedelta(days=1)).strftime("%Y-%m-%d")

    # Tentukan nama file untuk tanggal tersebut
    filename = f'temp_{since_date}.csv'

    # Simpan nama file ke dalam list
    file_list.append(filename)

    # Buat query pencarian untuk tanggal tersebut
    search_keyword = f'game slot lang:id until:{until_date} since:{since_date}'

    # Jalankan perintah tweet-harvest untuk scraping
    command = f'npx -y tweet-harvest@2.6.1 -o "{filename}" -s "{search_keyword}" --tab "LATEST" -l {limit} --token {twitter_auth_token}'
    print(f'Running command: {command}')
    subprocess.run(command, shell=True)

    print(f'Successfully saved data for date {since_date} in {filename}')

    # Lanjut ke tanggal berikutnya
    current_dt += timedelta(days=1)
```

```python
import os
import pandas as pd
import zipfile

# Folder path tempat file CSV berada
folder_path = 'tweets-data'

# Ambil semua file CSV di folder_path
file_list = [os.path.join(folder_path, file) for file in os.listdir(folder_path) if file.endswith('.csv')]

# Nama file output final
final_filename = 'gameslot-mei.csv'
zip_filename = 'gameslot-mei.zip'

# List untuk menampung DataFrame valid
valid_dataframes = []

# Validasi setiap file CSV dan tambahkan ke list jika valid
for file in file_list:
    try:
        df = pd.read_csv(file)
        if not df.empty:
            valid_dataframes.append(df)
            print(f"{file} loaded successfully with {len(df)} rows.")
        else:
            print(f"{file} is empty. Skipping.")
    except pd.errors.EmptyDataError:
        print(f"{file} is empty or has no columns. Skipping.")

# Gabungkan semua DataFrame yang valid jika ada
if valid_dataframes:
    combined_df = pd.concat(valid_dataframes, ignore_index=True)
    combined_df.to_csv(final_filename, index=False)
    print(f"All data combined into {final_filename}")

    # Buat file zip untuk menyimpan file sementara
    with zipfile.ZipFile(zip_filename, 'w') as zipf:
        for file in file_list:
            zipf.write(file, os.path.basename(file))
            print(f"{file} added to {zip_filename}")

    # Hapus file sementara setelah dimasukkan ke dalam zip
    for file in file_list:
        os.remove(file)
        print(f"Temporary file {file} removed.")
else:
    print("No valid files to combine.")
```
